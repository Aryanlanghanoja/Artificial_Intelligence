```python
# 2. Import Required Libraries

import numpy as np
import pandas as pd
from surprise import SVD, Dataset, Reader
from surprise.model_selection import cross_validate
```

```python
# 3. Prepare the Sample Dataset

# Define users, items, and ratings
users = [1, 2, 3, 4, 1, 2, 3, 4]
movies = [
    "Star Wars",
    "Hary Porter",
    "Star Wars",
    "Star Wars",
    "Hary Porter",
    "Tom Rider",
    "Hary Porter",
    "Tom Rider",
]
ratings = [1, 3, 4, 2, 3, 4, 1, 1]

# Create a dictionary and convert to DataFrame
rating_dict = {
    "userID": users,
    "ItemID": movies,
    "rating": ratings
}

df = pd.DataFrame(rating_dict)
df
```

```python
# 4. Define Reader and Load Dataset

# Define the rating scale (min=1, max=5)
reader = Reader(rating_scale=(1, 5))

# Load dataset in Surprise format
data = Dataset.load_from_df(df[["userID", "ItemID", "rating"]], reader)
```

```python
# 5. Apply SVD Collaborative Filtering Algorithm

# Initialize the SVD algorithm
algo = SVD()

# Fit the algorithm on the data
algo.fit(data.build_full_trainset())
```

```python
# 6. Evaluate the Model using Cross Validation

cross_validate(algo , data , measures = ['rmse' , 'mae'] , cv  =5 , verbose = True)
```
